```
import os
import re
import sys
import json
import logging
from tqdm import tqdm

import cv2
import nibabel as nib
from PIL import Image

import pandas as pd
import numpy as np

## Pytorch packages
import torch

import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import DataLoader, Dataset, random_split
from torch.utils.tensorboard import SummaryWriter

from torch.optim.lr_scheduler import StepLR
import torch.optim as optim

from torchsummary import summary

from torchvision import datasets, transforms, models

## MONAI
import monai
from monai.apps import download_and_extract
from monai.config import print_config
from monai.data import DataLoader, ImageDataset
from monai.transforms import (
    EnsureChannelFirst,
    Compose,
    RandRotate90,
    Resize,
    ScaleIntensity,
    SignalFillEmpty,
)

from monai.visualize import plot_2d_or_3d_image
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    classification_report,
    confusion_matrix,
)

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
pin_memory = torch.cuda.is_available()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

logging.basicConfig(stream=sys.stdout, level=logging.INFO)
print_config()
```

```
# Image preprocessing
def standardize(input_array:np.array):
  input_array = np.nan_to_num(input_array, copy=False,)
  return (input_array - np.mean(input_array)) / np.std(input_array)

def normalize(input_array:np.array): # Min-Max Normalization
  input_array = np.nan_to_num(input_array, copy=False,)
  return (input_array - np.min(input_array)) / (np.max(input_array) - np.min(input_array)) * 255.

# Image file loading
def read_NIBImage (input_path:str):
  image = nib.load(input_path).get_fdata()
  # image = standardize(image)
  image = normalize(image)
  return image
```

```
# Paths
str_Workspace = os.getcwd()

# Path related
str_DatasetPath = str_ADPath

# Hyperparameters
int_Epochs = 50
int_BatchSize = 15

int_ImageSize = 96
win_size = (int_ImageSize, int_ImageSize, int_ImageSize)

float_LearningRate= 0.01
float_TestRatio = 0.2

# int_RandSeedModel = 42
int_RandSeedData = 5
```

```
# Path related
str_ImagesPath = os.path.join(str_DatasetPath, 'Images')
str_TrainDatasetPath = os.path.join(str_ImagesPath, 'Train')
str_TestDatasetPath = os.path.join(str_ImagesPath, 'Test')
```

```
df_train = pd.read_csv(os.path.join(str_DatasetPath, 'PatientInfo_Train.csv'))
df_test = pd.read_csv(os.path.join(str_DatasetPath, 'PatientInfo_Test.csv'))

df_train = df_train[df_train['Label'] != 'MCI']
df_test = df_test[df_test['Label'] != 'MCI']
df_train = df_train.reset_index().drop(columns=['index'])
df_test = df_test.reset_index().drop(columns=['index'])

_list_Label = list(df_train['Label'].unique())
_list_Label.sort(reverse=True)

dict_Label2Num = {_label:_num for _num,_label in enumerate(_list_Label)}
dict_Num2Label = {_num:_label for _num,_label in enumerate(_list_Label)}
print(dict_Label2Num, dict_Num2Label,sep='\n')
del _list_Label

df_train['Label_y'] = df_train['Label'].map(dict_Label2Num)
df_test['Label_y'] = df_test['Label'].map(dict_Label2Num)

print('Train:', df_train.shape)
print(df_train.groupby(by=['Label','Sex']).count()['Subject_ID'])
print('Test:', df_test.shape)
print(df_test.groupby(by=['Label','Sex']).count()['Subject_ID'])
df_train.head()
```

```
from sklearn.model_selection import train_test_split

df_train, df_valid = train_test_split(
    df_train,
    stratify = df_train[['Label','Sex']],
    test_size = float_TestRatio,
    random_state = int_RandSeedData,
)
df_train = df_train.reset_index().drop(columns=['index'])
df_valid = df_valid.reset_index().drop(columns=['index'])
```

```
print('Train:', df_train.shape)
print(df_train.groupby(by=['Label','Sex']).count()['Subject_ID'])
print('Valid:', df_valid.shape)
print(df_valid.groupby(by=['Label','Sex']).count()['Subject_ID'])
print('Test:', df_test.shape)
print(df_test.groupby(by=['Label','Sex']).count()['Subject_ID'])
df_train.head()
```

```
# Represent labels in one-hot format for binary classifier training,
# BCEWithLogitsLoss requires target to have same shape as input
Onehot_trainLabel = torch.nn.functional.one_hot(torch.as_tensor(df_train['Label_y'].tolist())).float()
Onehot_validLabel = torch.nn.functional.one_hot(torch.as_tensor(df_valid['Label_y'].tolist())).float()
Onehot_testLabel = torch.nn.functional.one_hot(torch.as_tensor(df_test['Label_y'].tolist())).float()
```

```
fig, ax = plt.subplots(nrows=5, ncols=6, figsize=(10,15))

for index in range(5):
    _file = df_train.loc[index, 'Images_Files']
    _imagePath = os.path.join(str_TrainDatasetPath, _file)
    image = read_NIBImage(_imagePath)
    # print(image.max(),image.min())
    # print(image.shape)
    _step = int((image.shape[2]-50)/6)

    ax[index][2].set_title(f"Label: {df_train.loc[index, 'Label']}",
            fontsize=15, fontweight ="bold")
    for i in range(6):
        ax[index, i].imshow(image[:, :, 30+_step*i], cmap='gray')
        ax[index, i].axis('off')
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```
# Define transforms
train_transforms = Compose([
    SignalFillEmpty(),
    ScaleIntensity(),
    EnsureChannelFirst(),
    Resize(win_size),
    RandRotate90(),
])

val_transforms = Compose([
    SignalFillEmpty(),
    ScaleIntensity(),
    EnsureChannelFirst(),
    Resize(win_size)
])

# Define nifti dataset, data loader
_str_ImagePaths = df_train['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
# print(myFun.read_NIBImage(_str_ImagePaths[0]))
check_ds = ImageDataset(image_files= _str_ImagePaths,
                        labels=Onehot_trainLabel,
                        reader= "NibabelReader",
                        transform=train_transforms)
check_loader = DataLoader(check_ds,
                          batch_size=5,
                          num_workers=5,
                          pin_memory=pin_memory)

im, label = monai.utils.misc.first(check_loader)
print(type(im), im.shape, label, label.shape)
# print(im)

# create a training data loader
_str_ImagePaths = df_train['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
train_ds = ImageDataset(image_files=_str_ImagePaths,
                        labels=Onehot_trainLabel,
                        reader= "NibabelReader",
                        transform=train_transforms)
train_loader = DataLoader(train_ds,
                          batch_size= int_BatchSize,
                          shuffle=True,
                          num_workers=int_BatchSize,
                          pin_memory=pin_memory)

# create a validation data loader
_str_ImagePaths = df_valid['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
val_ds = ImageDataset(image_files=_str_ImagePaths,
                      labels=Onehot_validLabel,
                      reader= "NibabelReader",
                      transform=val_transforms)
val_loader = DataLoader(val_ds,
                        batch_size=int_BatchSize,
                        num_workers=int_BatchSize,
                        pin_memory=pin_memory)

# create a testing data loader
_str_ImagePaths = df_test['Images_Files'].apply(lambda x:os.path.join(str_TestDatasetPath,x))
test_ds = ImageDataset(image_files=_str_ImagePaths,
                      labels=Onehot_testLabel,
                      reader= "NibabelReader",
                      transform=val_transforms)
test_loader = DataLoader(test_ds,
                        batch_size=int_BatchSize,
                        num_workers=int_BatchSize,
                        pin_memory=pin_memory)
```

```
# Create DenseNet121, CrossEntropyLoss and Adam optimizer
NetArch = monai.networks.nets.DenseNet121(
    spatial_dims=3,
    in_channels=1,
    out_channels=len(dict_Label2Num.keys()),
)
model=NetArch.to(device)

# loss_function = torch.nn.CrossEntropyLoss()
loss_function = torch.nn.BCEWithLogitsLoss()  # also works with this data

optimizer = torch.optim.Adam(model.parameters(), float_LearningRate)
```

```
# start a typical PyTorch training
val_interval = 1
best_metric = -1
best_metric_epoch = -1
epoch_loss_values = []
metric_values = []
# scaler = torch.cuda.amp.GradScaler()
writer = SummaryWriter()

for epoch in range(int_Epochs):
    print("-" * 10)
    print(f"epoch {epoch + 1}/{int_Epochs}")
    model.train()
    epoch_loss = 0
    step = 0

    for batch_data in train_loader:
        step += 1
        inputs, labels = batch_data[0].to(device), batch_data[1].to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_function(outputs, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        epoch_len = len(train_ds) // train_loader.batch_size
        print(f"{step}/{epoch_len}, train_loss: {loss.item():.4f}")
        writer.add_scalar("train_loss", loss.item(), epoch_len * epoch + step)

    epoch_loss /= step
    epoch_loss_values.append(epoch_loss)
    print(f"epoch {epoch + 1} average loss: {epoch_loss:.4f}")

    if (epoch + 1) % val_interval == 0:
        model.eval()

        num_correct = 0.0
        metric_count = 0
        # for val_data in val_loader:
        for val_data in val_loader:
            val_images, val_labels = val_data[0].to(device), val_data[1].to(device)
            with torch.no_grad():
                val_outputs = model(val_images)
                value = torch.eq(val_outputs.argmax(dim=1), val_labels.argmax(dim=1))
                metric_count += len(value)
                num_correct += value.sum().item()

        metric = num_correct / metric_count
        metric_values.append(metric)

        if metric >= best_metric:
            best_metric = metric
            best_metric_epoch = epoch + 1
            torch.save(model.state_dict(), "best_metric_model_classification3d_array(AD).pth")
            print("saved new best metric model")

        print(f"Current epoch: {epoch+1} current accuracy: {metric:.4f} ")
        print(f"Best accuracy: {best_metric:.4f} at epoch {best_metric_epoch}")
        writer.add_scalar("val_accuracy", metric, epoch + 1)

print(f"Training completed, best_metric: {best_metric:.4f} at epoch: {best_metric_epoch}")
writer.close()
```

```
plt.plot(epoch_loss_values, label="training loss")
val_epochs = np.linspace(1, int_Epochs, np.floor(int_Epochs / val_interval).astype(np.int32))
plt.plot(val_epochs, metric_values, label="validation acc")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.show()
```

```
# # Reload the best network and display info
# del model
# model=NetArch.to(device)
# model.load_state_dict(torch.load("best_metric_model_classification3d_array(ASD).pth"))
# model.eval()

y_pred = torch.tensor([], dtype=torch.float32, device=device)
y = torch.tensor([], dtype=torch.long, device=device)

for val_data in test_loader:
    val_images = val_data[0].to(device)
    val_labels = val_data[1].to(device).argmax(dim=1)

    outputs = model(val_images)
    y_pred = torch.cat([y_pred, outputs.argmax(dim=1)], dim=0)
    y = torch.cat([y, val_labels], dim=0)

print(
    classification_report(
        y.cpu().numpy(),
        y_pred.cpu().numpy(),
        target_names=dict_Label2Num.keys(),
    )
)

cm = confusion_matrix(
    y.cpu().numpy(),
    y_pred.cpu().numpy(),
    normalize="true",
)
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=dict_Label2Num.keys()
)
disp.plot(ax=plt.subplots(1, 1, facecolor="white")[1])
plt.show()
```

```
# cam = monai.visualize.CAM(nn_module=model, target_layers="class_layers.relu", fc_layers="class_layers.out")
cam = monai.visualize.GradCAM(nn_module=model, target_layers="class_layers.relu")
# cam = monai.visualize.GradCAMpp(nn_module=model, target_layers="class_layers.relu")
print(
    "original feature shape",
    cam.feature_map_size([1, 1] + list(win_size), device),
)
print("upsampled feature shape", [1, 1] + list(win_size))

occ_sens = monai.visualize.OcclusionSensitivity(nn_module=model, mask_size=12, n_batch=1)

# For occlusion sensitivity, inference must be run many times. Hence, we can use a
# bounding box to limit it to a 2D plane of interest (z=the_slice) where each of
# the arguments are the min and max for each of the spatial dimensions (in this case HWD).
the_slice = test_ds[0][0].shape[-1] // 2
# the_slice = 60
occ_sens_b_box = [-1, -1, -1, -1, the_slice - 1, the_slice]
```

```
train_transforms.set_random_state(5)
n_examples = 10
subplot_shape = [3, n_examples]
plt.figure()
fig, axes = plt.subplots(*subplot_shape, figsize=(50, 15), facecolor="white")
items = np.random.choice(len(test_ds), size=len(test_ds), replace=False)

example = 0
for item in items:
    data = test_ds[item]  # this fetches training data with random augmentations
    image, label = data[0].to(device).unsqueeze(0), data[1][1]
    y_pred = model(image)
    pred_label = y_pred.argmax(1).item()
    # Only display tumours images
    if label == dict_Label2Num['Normal'] or label != pred_label:
        continue

    img = image.detach().cpu().numpy()[..., the_slice]
    # print(item,y_pred,label,data[1])
    name = "actual: "
    name += f"{dict_Num2Label[1]}" if label == 1 else f"{dict_Num2Label[0]}"
    name += "\npred: "
    name += f"{dict_Num2Label[1]}" if pred_label == 1 else f"{dict_Num2Label[0]}"
    name += f"\n{dict_Num2Label[1]}: {y_pred[0,1]:.3}"
    name += f"\n{dict_Num2Label[0]}: {y_pred[0,0]:.3}"

    # run CAM
    cam_result = cam(x=image, class_idx=None)
    cam_result = cam_result[..., the_slice]

    # run occlusion
    occ_result, _ = occ_sens(x=image, b_box=occ_sens_b_box)
    occ_result = occ_result[0, pred_label][None, None, ..., -1]

    for row, (im, title) in enumerate(
        zip(
            [img, cam_result, occ_result],
            [name, "CAM", "Occ. sens."],
        )
    ):
        cmap = "gray" if row == 0 else "jet"
        ax = axes[row, example]
        if isinstance(im, torch.Tensor):
            im = im.cpu().detach()
        im_show = ax.imshow(im[0][0], cmap=cmap)

        ax.set_title(title, fontsize=25)
        ax.axis("off")
        fig.colorbar(im_show, ax=ax)

    example += 1
    if example == n_examples:
        break
plt.show()
```
